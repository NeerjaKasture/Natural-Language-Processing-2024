```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import TimeoutException
import pandas as pd
import time

urls = ['https://urdu.siasat.com/category/india/page/']

topics = 2
link = []
start_page = 401  # Start extracting from page 301
total_pages = 2607  # Specify the total number of pages to scrape (you can adjust this as needed)

for base_url in urls:
    try:
        chrome_options = Options()
        chrome_options.add_argument("--headless")  # Headless mode for Colab
        chrome_options.add_argument("--no-sandbox")
        chrome_options.add_argument("--disable-dev-shm-usage")
        driver = webdriver.Chrome(options=chrome_options)
        print("Running in headless mode.")
        df = pd.DataFrame(columns=["url"])
        # Loop through the pages starting from start_page
        for pages in range(start_page, total_pages + 1):
            current_url = f"{base_url}{pages}/"
            driver.get(current_url)

            try:
                WebDriverWait(driver, 15).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'main.site-main')))
                div1 = driver.find_elements(By.CSS_SELECTOR, 'main.site-main')

                if len(div1) != 1:
                    print(f"Some problem with link: {current_url}")
                    break

                article = div1[0].find_elements(By.TAG_NAME, "article")
                if len(article) < 1:
                    print(f"Some problem with articles on link: {current_url}")
                    break

                # Extract URLs from articles
                for i in range(len(article)):
                    url = article[i].find_element(By.TAG_NAME, "a").get_attribute("href")
                    link.append(url)

                # Store the links in the dataframe
                for i in range(len(link)):
                    df.loc[len(df)] = [link[i]]

                print("Page", pages)

                # Save every 100 pages
                if pages % 100 == 0:
                    df = pd.DataFrame(link, columns=["url"])  # Create DataFrame from the list
                    df.to_csv(f"siasat_topic_{topics}_{pages}_url.csv", index=False)

            except TimeoutException:
                print(f"Timeout waiting for page elements: {current_url}")
                break
            except Exception as e:
                print(f"Error processing the page: {e}")
                break

    except Exception as e:
        print(f"Error initiating driver: {e}")

    finally:
        driver.quit()

# Save the collected links to a CSV file after extraction is complete
pd.DataFrame(link, columns=["url"]).to_csv("siasat_topic_2_urls.csv", index=False)
```

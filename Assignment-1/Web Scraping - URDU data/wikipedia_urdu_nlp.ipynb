```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Define the URL of the featured articles Wikipedia page
base_url = 'https://ur.wikipedia.org/wiki/%D9%88%DB%8C%DA%A9%DB%8C%D9%BE%DB%8C%DA%88%DB%8C%D8%A7:%D9%85%D9%86%D8%AA%D8%AE%D8%A8_%D9%85%D8%B6%D8%A7%D9%85%DB%8C%D9%86'

# Send a request to fetch the page content
response = requests.get(base_url)
soup = BeautifulSoup(response.text, 'html.parser')

# Initialize lists to store data
titles = []
urls = []
contents = []

# Counter to limit scraping to the first two articles
article_count = 0
max_articles = 10
# Step 1: Find and extract article links under the specified <p> tags
for div in soup.find_all('div', {'class': 'mw-content-rtl mw-parser-output'}):
    for p_tag in div.find_all('p'):
        for a_tag in p_tag.find_all('a', href=True):
            # Break the loop if we've scraped the desired number of articles
            if article_count >= max_articles:
                break
            # Extract the article URL
            # article_url = 'https://ur.wikipedia.org' + a_tag['href']
            if a_tag['href'].startswith('https://'):
                article_url = a_tag['href']
            else:
                article_url = 'https://en.wikipedia.org' + a_tag['href']

            # Fetch the article content
            article_response = requests.get(article_url)
            article_soup = BeautifulSoup(article_response.text, 'html.parser')

            # Extract the title of the article
            title_tag = article_soup.find('h1', {'class': 'firstHeading mw-first-heading'})
            title = title_tag.get_text(strip=True) if title_tag else 'No title found'

            # Extract the content of the article (from <p> tags)
            content_div = article_soup.find('div', {'class': 'mw-content-rtl mw-parser-output'})
            content = ''
            if content_div:
                for p in content_div.find_all('p'):
                    content += p.get_text(strip=True)

            # Append the data to the lists
            titles.append(title)
            urls.append(article_url)
            contents.append(content)

            # Increment article counter
            article_count += 1

        # Break the loop if we've scraped the desired number of articles
        if article_count >= max_articles:
            break
    # Break the loop if we've scraped the desired number of articles
    if article_count >= max_articles:
        break

# Step 2: Create a DataFrame from the lists
df = pd.DataFrame({
    'Title': titles,
    'URL': urls,
    'Content': contents
})

# Step 3: Save DataFrame to CSV
df.to_csv('wikipedia_urdu.csv', index=False)
print("Scraping completed")

# df = pd.read_csv('/content/wikipedia_urdu.csv')
# df
```
